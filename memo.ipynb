__Variance__ of a variable describes how much the values are spread <br>
average of the squared differences of the mean. <br>
higher the variance, data are spread out from the mean<br>
$V(X) = \frac{1}{n} \sum \limits_{i = 1}^n(x_i -\bar{x})^2$

__Standard Deviation__ shows how far each value likes from the means <br>
square root of the variance

__Covariance__ is a measure that tells the amout of dependency between two variables <br>
positive, negative, 0 (not dependent) <br>
$Cov(X,Y) = \frac{1}{n} \sum \limits_{i = 1}^n(x_i -\bar{x})(y_i -\bar{y})$

__Correlation__ normalized covariance


```python
import math
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots

from ipywidgets import interact
import ipywidgets as widgets
```

```python
lag = np.arange(1,500,1) #lag of correlation
t = np.arange(0,1, 0.001)
y = np.sin(2*np.pi*10*t)
N = 500 #should be samller than t.size-lag

#yn = np.pad(y, (0, lag.size), 'constant')
    
r = np.zeros(lag.size)
for j in range (lag.size):
    a = 0
    for i in range(N):
        a = a + y[i]*y[i+j]
    r[j] = a/N

A = np.correlate(y, y, 'full')/500
    
# plt.plot(lag,r)
# plt.plot(y)
```
